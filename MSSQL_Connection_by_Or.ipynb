```python
FRAUD_age=%sql SELECT count(*) total_customer, * FROM (SELECT case   when Age <18 then 'Under 18'   when Age between 18 and 30 then '18-30'   when Age between 31 and 50 then '31-50'   when Age between 51 and 65 then '51-65'   when  Age>65 then '>65' END as Age FROM [AntiLaundering].[dbo].[FraudDetectedTrend] where [SUSPECTFRAUD]=1 ) t  group by Age order by count(*)
FRAUD_age
```

```python
df_FRAUD_age = pd.DataFrame(FRAUD_age,columns=['total_customer', 'Age'])
df_FRAUD_age
```

```python
Age = df_FRAUD_age['Age']
total_customer = df_FRAUD_age['total_customer']
 
# Figure Size
fig = pyplt.figure(figsize =(35, 10))
 
# Horizontal Bar Plot
pyplt.bar(Age, total_customer)
 
# Show Plot
pyplt.show()
```
