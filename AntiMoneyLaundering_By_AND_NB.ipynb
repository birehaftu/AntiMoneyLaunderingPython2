```python
from sklearn.tree import DecisionTreeClassifier # Import Decision Tree Classifier
from sklearn.model_selection import train_test_split # Import train_test_split function
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation
```

```python
col_names = ['TXN_ID','TOTAL_TNX','SUSPECTFRAUD','DEBIT_CUSTOMER','USD_EQUIV_BALANCE','CREDIT_AMOUNT','AGE','GENDER','REGIONNAME','DISTRICTNAME']
# load dataset
```

```python
sql_allData=%sql SELECT  [TXN_ID]      ,[TOTAL_TNX]      ,[SUSPECTFRAUD]      ,[DEBIT_CUSTOMER]      ,[USD_EQUIV_BALANCE]      ,[CREDIT_AMOUNT]      ,[AGE]      ,[GENDER]      ,[REGIONNAME]      ,[DISTRICTNAME]  FROM [AntiLaundering].[dbo].[FraudDetectedTrendAND]
df_Fraud_All = pd.DataFrame(sql_allData,columns=col_names)
#df_Fraud_All
df_Fraud_All.head(10)
```

```python
gender = {'MALE': 0,'FEMALE': 1}
district = {'ADAMA':0,'AMBO':1,'ARADA':2,'ASSELA':3,'BAHIR DAR':4,'BALE ROBE':5,'BOLE':6,'CONSUMER EAST ADDIS':7,'CONSUMER NORTH ADDIS':8,'CONSUMER SOUTH ADDIS':9,'CONSUMER WEST ADDIS':10,'DEBRE BREHAN':11,'DEBRE MARKOS':12,'DESSIE':13,'DILLA':14,'DIRE DAWA':15,'GONDER':16,'HAWASSA':17,'HOSSANA':18,'JIJIGA':19,'JIMMA':20,'KIRKOS':21,'KOLFE':22,'MEGENAGNA':23,'MEKELLE':24,'MERKATO':25,'METTU':26,'NEKEMTE':27,'NIFAS SILK':28,'SEMERA':29,'SHASHEMENE':30,'SHIRE':31,'WOLAITA SODO':32,'WOLDIA':33,'YEKA':34}
region={'ADDIS ABABA':0,'AFAR':1,'AMHARA':2,'BENSHANGUL GUMUZ':3,'DIRE DAWA':4,'GAMBELLA':5,'HARARI':6,'OROMIYA':7,'SIDAMA':8,'SNNP':9,'SOMALE':10,'SWER':11,'TIGRAY':12}

 # traversing through dataframe
# Gender column and writing
# values where key matches
df_Fraud_All_raverse=df_Fraud_All.copy()
df_Fraud_All_raverse.GENDER = [gender[item] for item in df_Fraud_All_raverse.GENDER]
df_Fraud_All_raverse.REGIONNAME = [region[item] for item in df_Fraud_All_raverse.REGIONNAME]
df_Fraud_All_raverse.DISTRICTNAME = [district[item] for item in df_Fraud_All_raverse.DISTRICTNAME]
df_Fraud_All_raverse.head(10)
```

```python
#split dataset in features and target variable
feature_cols = ['TOTAL_TNX','DEBIT_CUSTOMER','USD_EQUIV_BALANCE','CREDIT_AMOUNT','AGE','GENDER','REGIONNAME','DISTRICTNAME']
 
X = df_Fraud_All_raverse[feature_cols] # Features
y = df_Fraud_All_raverse.SUSPECTFRAUD # Target variable
```

```python
# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1) # 70% training and 30% test
#X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1) # 80% training and 20% test
```

```python
# neural network with keras 
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

x_train_ = np.asarray(X_train).astype(np.float32)
y_train_ = np.asarray(y_train).astype(np.float32)
# define the keras model
model = Sequential()
model.add(Dense(1, activation='sigmoid'))
# compile the keras model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
# fit the keras model on the dataset
model.fit(x_train_, y_train_, epochs=15, batch_size=10)
# evaluate the keras model
_, accuracy = model.evaluate(x_train_, y_train_)
print('Accuracy: %.2f' % (accuracy*100))
```
